```python
import os

import numpy as np
import pandas as pd
import cv2
from PIL import Image
import PIL.ImageOps
from sklearn.model_selection import train_test_split
from sklearn import datasets
import tensorflow as tf

import matplotlib.pyplot as plt
import seaborn as sns


import cv2

from bs4 import BeautifulSoup as BSHTML
import urllib
import requests

from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from sklearn import preprocessing
#ray.init()
```

```python
files = ['fashionmnist','coil-100','paris_buildings', 'pets', 
         'oxford_buildings','dogs','cinic10', 'caltech',
         'flowers','paintings']

# files = ['caltech','paris_buildings', 'pets', 
#          'oxford_buildings','dogs',
#          'flowers','paintings']

def getImage(path):
    return OffsetImage(plt.imread(path))


#@ray.remote
def plot_projection(f, cp, csv, labels, paths,idx):
    print(csv)
    try:
        df = pd.read_csv(orig_path + csv)
    except: 
        return

    df = df.loc[idx,:]
    df['paths'] = paths
    df['ll'] = labels

    fig, ax = plt.subplots(figsize=(100,100))
    ax.scatter(df.iloc[:,0], df.iloc[:,1]) 
    for i,row in df.iterrows():
        try:
            #print(row['paths'])
            ab = AnnotationBbox(getImage(row['paths']), (row[0], row[1]), frameon=True)
        except:
            try:
                ab = AnnotationBbox(getImage(row['paths'].split('.png')[0]+'.jpg'), (row[0], row[1]), frameon=True)
            except:
                continue
        ab.patch.set_linewidth(4)
        ab.patch.set_edgecolor(cp[int(row['ll']) if row['ll']<11 else 11])
        ab.patch.set_facecolor(cp[int(row['ll']) if row['ll']<11 else 11])
        ax.add_artist(ab)

    print(f+'/figures200/img_'+csv.split('.csv')[0]+'.pdf')
    plt.savefig(f+'/figures200/img_'+csv.split('.csv')[0]+'.pdf')
    plt.close()


cp = sns.color_palette("Paired")

for f in files:
    orig_path = f + '/data200/'
    
    try:
        ddf = pd.read_csv(f + '/data200/PCA.csv')
    except:
        continue
        
    if len(ddf)>400:
        _, ddf = train_test_split(ddf, test_size=400, stratify=ddf['labels'])
    
    idx = ddf.index
    
    paths = list(f+'/thumbnails200/'+ ddf['filename']+'.png')
    le = preprocessing.LabelEncoder()
    labels = le.fit_transform(ddf['labels'])

    
    print(f)
    for csv in os.listdir(orig_path):
        if csv.endswith('.csv'):
            if os.path.exists(f+'/figures200/img_'+csv.split('.csv')[0]+'.pdf'):
                #print('.')
                continue 
            plot_projection(f, cp, csv, labels, paths,idx)
                
            
            

            
            
```
